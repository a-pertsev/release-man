```python
def get_value(some_skills, some_project_skills):
    result = 0
    for i in xrange(some_project_skills.size):
        if some_project_skills[i] != 0:
           result += min(some_skills[i], some_project_skills[i])
    return result
```

```python
def levels_reachable(level, money):
    ups = 0
    for i in xrange(100):
        cost = level + ups + 1
        money -= cost
        if money > 0:
            ups +=1 
    return ups
```

```python
def print_matrix(matrix):
    print "="*30
    for row in matrix:
        print row
    print "="*30


def calc_guy_price(project_skills, guy, team, people_data):
#     if len(team) > 2:
#         print 'calc guy price ', people_data[guy]['name']
#         print 'project ', project_skills, '\n', 'guy skills ', people_data[guy]['skills'], '\n', team

    team_project_skills = project_skills
    for man in team:
        team_project_skills = team_project_skills - people_data[man]['skills']


    guy_skills = people_data[guy]['skills']

    total_matches = 0

    for skill_idx, skill_level in enumerate(team_project_skills):
        total_matches += max(0, min(guy_skills[skill_idx], skill_level))

    # for skill in project_skills:
    #     for guy_spec in guy['specs']:
    #         if guy_spec['name'] == skill['name']:
    #             total_matches += max(0, min(guy_spec['level'], skill['required']))

#     if len(team) > 2:
#         print project_skills, '\n', total_matches
    return total_matches


def find_guy_ans(things_to_take, current_weight, answer, cost_matrix):
    if cost_matrix[things_to_take][current_weight] == 0:
        return
    if cost_matrix[things_to_take-1][current_weight] == cost_matrix[things_to_take][current_weight]:
        find_guy_ans(things_to_take-1, current_weight, answer, cost_matrix)
    else:
        find_guy_ans(things_to_take-1, current_weight-1, answer, cost_matrix)
        answer.append(things_to_take-1)


def calc_team_need_skills(project_skills, team, people_data):
    team_project_skills = project_skills
    for man in team:
        team_project_skills = team_project_skills - people_data[man]['skills'] 

    team_sum = 0
    for value in team_project_skills:
        if value > 0:
            team_sum += value

    for man in team:
        team_sum -= people_data[man]['level_ups']
        
    return team_sum

def dynamic_hire(project_skills, need_to_hire_num, people_data, filtered_guys):
    cost_matrix = []

    for i in xrange(0, len(filtered_guys)+1):
        cost_matrix.append([None for j in  xrange(0, need_to_hire_num+1)])

    cost_matrix[0] = [0 for  i in xrange(0, need_to_hire_num+1)]

    for i in xrange(0, len(filtered_guys)+1):
        cost_matrix[i][0] = 0

    guys_matrix = [[[] for j in xrange(0, need_to_hire_num+1)] for i in xrange(0, len(filtered_guys)+1)]

    for guy_idx_to_take in xrange(1, len(filtered_guys)+1):
        for current_project_size in xrange(1, need_to_hire_num+1):
            guy = filtered_guys[guy_idx_to_take-1]
            if current_project_size >= 1:
                price_with_guy = cost_matrix[guy_idx_to_take-1][current_project_size - 1] +  calc_guy_price(project_skills, guy, guys_matrix[guy_idx_to_take-1][current_project_size - 1], people_data)
                price_without_guy = cost_matrix[guy_idx_to_take-1][current_project_size]
                guys_matrix[guy_idx_to_take][current_project_size].extend(guys_matrix[guy_idx_to_take-1][current_project_size - 1])
                if (price_with_guy > price_without_guy):
                    guys_matrix[guy_idx_to_take][current_project_size].append(guy)
                    cost_matrix[guy_idx_to_take][current_project_size] = price_with_guy
                else:
                    cost_matrix[guy_idx_to_take][current_project_size] = price_without_guy
            else:
                cost_matrix[guy_idx_to_take][current_project_size] = cost_matrix[guy_idx_to_take-1][current_project_size]

#     print_matrix(cost_matrix)
#     print_matrix(guys_matrix)
    guys_from_usefull = []
    find_guy_ans(len(cost_matrix)-1, len(cost_matrix[0])-1, guys_from_usefull, cost_matrix)
    return [filtered_guys[i] for i in guys_from_usefull]
```
